```python
# importing libraries to read data
```

```python
import numpy as np
import pandas as pd
```

```python
# fetching the dataset
```

```python
df = pd.read_csv('Grades.csv')

df
```

```python
# checking the shape of dataset and its datatype also check the null values in dataset
```

```python
# some null values are there in dataset, let's drop these values
```

```python
df = df.dropna()
```

```python
df.shape
```

```python
df.sample(n=5)
```

```python
# now describe the data
```

```python
df.describe()
```

```python
# all features columns are in object as we seen before let's deal with these columns by using get_dummies method
```

```python
numerics = ['int8','int16','int32','int64','float16','float32','float64']

categorical_columns = []

features = df.columns.values.tolist()

for col in features:
    if df[col].dtype in numerics:
        continue
    categorical_columns.append(col)
```

```python
categorical_columns
```

```python
len(categorical_columns)
```

```python
df_dummies = pd.get_dummies(df[categorical_columns],drop_first=True)
```

```python
df_dummies.head()
```

```python
df_dummies.shape
```

```python
df = df.join(df_dummies)
```

```python
df
```

```python
df.shape
```

```python
df.drop(columns=categorical_columns,axis=1,inplace=True)
```

```python
df
```

```python
# assign the data into two variables
```

```python
x = df.drop(columns=['CGPA'])

y = df['CGPA']
```

```python
# import tools to standardize the data and model to train the dataset
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

x_scaled = scaler.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_scaled,y,test_size=0.25,random_state=42)
```

```python
# linear regression
```

```python
lr = LinearRegression()

lr.fit(x_train,y_train)
```

```python
x_test_pred = lr.predict(x_test)
```

```python
lr.score(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

```python
x_test_pred
```

```python
# adaboost regressor
```

```python
from sklearn.ensemble import AdaBoostRegressor
```

```python
ada = AdaBoostRegressor()

ada.fit(x_train,y_train)
```

```python
x_test_pred1 = ada.predict(x_test)
```

```python
ada.score(x_train,y_train)
```

```python
ada.score(x_test,y_test)
```

```python
x_test_pred1
```

```python
# hyperparameter tuning using randomized searchcv
```

```python
from sklearn import metrics
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
params = {'n_estimators':[45,50,55,60],'learning_rate':[0.20,0.25,0.30,0.35]}
```

```python
rnd_srch = RandomizedSearchCV(AdaBoostRegressor(),cv=5,param_distributions=params)
```

```python
rnd_srch.fit(x_train,y_train)
```

```python
rnd_srch.best_estimator_
```

```python
ada1 = AdaBoostRegressor(learning_rate=0.35,n_estimators=55)


ada1.fit(x_train,y_train)

x_test_pred2 = ada1.predict(x_test)


print('**Accuracy post tuning**')

print(metrics.r2_score(y_test,x_test_pred2))
```
