```python
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
plt.figure(figsize = (20,15))
plotnumber = 1

for column in df:
    if plotnumber<=10:
        ax = plt.subplot(2,5,plotnumber)
        sns.distplot(df[column])
        plt.xlabel(column,fontsize=20)
        
        
    plotnumber+=1    
plt.show() 
```

```python
# assign the data to variables
```

```python
x = df.drop(columns=['Happiness Score','Dystopia Residual'])

y = df['Happiness Score']
```

```python
x
```

```python
y
```

```python
plt.figure(figsize = (15,10))
plotnumber = 1

for column in x:
    if plotnumber<=10:
        ax = plt.subplot(2,5,plotnumber)
        plt.scatter(x[column],y)
        plt.xlabel(column,fontsize=10)
        plt.ylabel('Happiness Score',fontsize=10)
    plotnumber+=1
    
plt.show()
```

```python
# standardise the data
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

x_scaled = scaler.fit_transform(x)
```

```python
x_scaled
```

```python
# train test split and importing the model and fit the data to model for prediction
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_scaled,y,test_size=0.25,random_state=42)
```

```python
x_train
```

```python
y_train
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
x1 = lr.predict(x_train)
x1
```

```python
lr.score(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

```python
x2 = lr.predict(x_test)
x2
```

```python
plt.scatter(y_test,x2)

plt.xlabel('Actual Happiness Score')
plt.ylabel('Predicted Happiness Score')
plt.title('Actual vs Predicted')

plt.show()
```

```python
# checking error of model
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
x2 = lr.predict(x_test)
```

```python
mean_absolute_error(y_test,x2)
```

```python
mean_squared_error(y_test,x2)
```
