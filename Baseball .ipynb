```python
# Now it's time process the data.

# First we standarize the data because different columns have different units, so after standarize all columns will be in standard form.

# Because it will be better to give standard data to Model.

# Then we will do train_test_split.

# After that, will start for model training with different-different algorithms.
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
scaler = StandardScaler()

x_scaled = scaler.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_scaled,y,test_size=0.30,random_state=100)
```

```python
# Because given-data-target is continous data, so we use regression model for this dataset.

# Assign model to a variable.

# Train the data.
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
# Checking adjusted r2, how much model get trained in given dataset.
```

```python
lr.score(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

```python
# For training Model is giving 100% but for prediction it becomes worst.
```

```python
# Now,let's try to predict.
```

```python
lr_pred = lr.predict(x_test)
```

```python
lr_pred
```

```python
# By plotting scatter graph let's see how model predicted the data.
```

```python
plt.scatter(y_test,lr_pred)
plt.xlabel('Actual Win')
plt.ylabel('Predicted Win')
plt.title('Actual vs Predicted')

plt.show()
```

```python
# By different metrics let's check, how much error model show.
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
mean_absolute_error(y_test,lr_pred)
```

```python
mean_squared_error(y_test,lr_pred)
```

```python
np.sqrt(mean_squared_error(y_test,lr_pred))
```
